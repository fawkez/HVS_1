```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp

# Differential equation for f(t) from equation (12)
def dfdt(t, f):
    return np.sqrt(2) / (1 + np.cos(f))**2

# Numerically integrate f(t) from t = -3 to t = 15
t_range = (-3, 15)
f0 = 0  # initial condition for f (assuming f(0) = 0)
sol = solve_ivp(dfdt, t_range, [f0], t_eval=np.linspace(-3, 15, 500))

# Extract time and f values
t_values = sol.t
f_values = sol.y[0]

# Define equations (13) to (18) using f(t)
def eq13(f):
    x = -np.sin(f) / (1 + np.cos(f))
    y = np.cos(f) / (1 + np.cos(f))
    return x, y

def eq14(f):
    x = -np.sin(f)
    y = 1 + np.cos(f)
    return x, y

def eq15(f):
    x = 2 - np.cos(f)
    y = -np.cos(f) * np.tan(f / 2)
    return x, y

def eq16(f):
    x = (8 + 12 * np.cos(f)) * np.tan(f / 2)**4
    y = (35 * np.sin(f) - 2 * np.sin(2 * f) + 3 * np.sin(3 * f)) / (1 + np.cos(f))**2
    return x, y

def eq17(f):
    z = 2 * np.sin(f) / (1 + np.cos(f))
    return z

def eq18(f):
    z = 2 * np.cos(f) / (1 + np.cos(f))
    return z

```

```python
# Plotting the equations
fig, axs = plt.subplots(3, 2, figsize=(10, 15))

# Plot for eq13
x13, y13 = eq13(f_values)
axs[0, 0].plot(t_values, x13, label='x')
axs[0, 0].plot(t_values, y13, label='y')
axs[0, 0].set_title('Equation (13)')
axs[0, 0].legend()

# Plot for eq14
x14, y14 = eq14(f_values)
axs[0, 1].plot(t_values, x14, label='x')
axs[0, 1].plot(t_values, y14, label='y')
axs[0, 1].set_title('Equation (14)')
axs[0, 1].legend()

# Plot for eq15
x15, y15 = eq15(f_values)
axs[1, 0].plot(t_values, x15, label='x')
axs[1, 0].plot(t_values, y15, label='y')
axs[1, 0].set_title('Equation (15)')
axs[1, 0].legend()

# Plot for eq16
x16, y16 = eq16(f_values)
axs[1, 1].plot(t_values, x16, label='x')
axs[1, 1].plot(t_values, y16, label='y')
axs[1, 1].set_title('Equation (16)')
axs[1, 1].legend()

# Plot for eq17
z17 = eq17(f_values)
axs[2, 0].plot(t_values, z17)
axs[2, 0].set_title('Equation (17)')

# Plot for eq18
z18 = eq18(f_values)
axs[2, 1].plot(t_values, z18)
axs[2, 1].set_title('Equation (18)')

plt.tight_layout()
plt.show()
```
